```python
# !pip install beautifultable
from beautifultable import BeautifulTable
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
myTxt = "Apple is looking at buying U.K. startup for $1 billion"
doc = nlp(myTxt)
```

```python
table = BeautifulTable()
```

```python
table.columns.header = ['Text', 'POS', 'TAG', 'Explaination TAG', 'Dep', 'Ent', 'Shape', 'is_alpha', 'is_stop']
for token in doc:
    table.rows.append([token.text, token.pos_, token.tag_, spacy.explain(token.tag_), token.dep_, token.ent_type_, token.shape_, token.is_alpha, token.is_stop ])
```

```python
print(table)
```

## Visualization

```python
from spacy import displacy
```

```python
# In python code we use:
#     - displacy.serve(doc, style='dep')
# In jupyter we use:
#     - displacy.render(doc, style='dep')
options = {"compact": True, "distance":100, "bg": "#385a7c", "color": "#b2eee6", "font": "Monospace"}
displacy.render(doc, style='dep', jupyter=True, options=options)
```

```python
displacy.render(doc, style='ent', jupyter=True)
```
