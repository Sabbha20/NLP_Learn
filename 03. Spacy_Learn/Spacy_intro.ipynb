```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm") # returns language object
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
docS = nlp("Sabbha is the most powerful person and multi-trillionaire. Hurray!!!")
docS.to_json()
```

### Tokenization

```python
for token in docS:
    print(f"Text:\t{token.text}\nPOS:\t{token.pos_}\nDep:\t{token.dep_}\nLemma:\t{token.lemma_}\n\n")
```
