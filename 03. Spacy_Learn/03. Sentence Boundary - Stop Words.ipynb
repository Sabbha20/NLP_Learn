## Sentence Boundary

```python
import spacy
from beautifultable import BeautifulTable
from spacy import displacy
```

```python
# DECLARING LANGUAGE MODEL
nlp = spacy.load("en_core_web_sm")
```

```python
mypara = """The iPhone, released in 2007, revolutionized the mobile phone industry with its 3.5-inch touchscreen, intuitive iOS operating system, 
and groundbreaking App Store, boasting over 2 million apps as of 2023. Its sleek design and cutting-edge features, like the 12MP camera 
and A17 Bionic chip, solidified its position as a status symbol and cultural icon, selling over 1 billion units worldwide. 
Despite facing fierce competition from Android devices, the iPhone remains a market leader, praised for its user-friendly interface, 
seamless integration with Apple's ecosystem, and consistent software updates, ensuring its continued relevance for years to come."""
```

```python
doc = nlp(mypara)
```

```python
doc.text
```

```python
doc
```

```python
sentences = list(doc.sents)
```

```python
for n, sentence in enumerate(sentences):
    print(f"{n+1}.\t {sentence}")
```

## Stop Words

```python
stop_words = spacy.lang.en.stop_words.STOP_WORDS
```

```python
stop_words # Sometimes keeping stop words helps you not always
```

### Filtering stop words from paragraph - mypara

```python
mypara
```

```python
for n, sentence in enumerate(sentences):
    print(sentence.text)
    for token in sentence:
        if token.text not in stop_words:
            print(token)
    break
```
